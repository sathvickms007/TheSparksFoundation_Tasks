```python
yearwise_attacks = df.groupby('Year').size().reset_index(name='count')
sb.lineplot(x='Year', y='count', data=yearwise_attacks, color='red')
plt.xlabel('Year')
plt.ylabel('Count of Attacks')
plt.title("Terrorist Attacks")
plt.show()
```

```python
yearwise_attacks.describe()
```

```python
#Observation:
#Maximum no. of attacks was recorded on the year 2017
```

```python
regions_attacked = df.groupby(['Year','Region']).size().reset_index(name='count')
plt.subplots(figsize=(15,6))
sb.lineplot(x='Year',y='count',hue='Region',data=regions_attacked)
plt.title('Region wise attacks')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sb.countplot(x='Region',data=df,palette='inferno',order=df['Region'].value_counts().index)
plt.xticks(rotation=80)
plt.title('count of Terrorist activities grouped by regions')
plt.show()
```

```python
groups = a[['attacktype1_txt','gname','weaptype1_txt','targtype1']]
group_counts = groups['gname'].value_counts()
sort = group_counts.sort_values(ascending=False)
sort = sort.iloc[1:]
top_5 = sort.head(5)


sb.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sb.barplot(x=top_5.index, y=top_5.values, palette="viridis")
plt.title('Top 5 Terrorist communities')
plt.xlabel('Attack Count')
plt.ylabel('Group')
plt.xticks(rotation = 90)
plt.show()
```

Final Analysis through conclusion:
